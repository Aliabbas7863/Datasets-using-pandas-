```python
import pandas as pd
import random
```

```python
# Programming languages with introduction years
languages = {
    "Python": 1991,
    "JavaScript": 1995,
    "Java": 1995,
    "C": 1972,
    "C++": 1985,
    "C#": 2000,
    "Go": 2009,
    "Rust": 2010,
    "TypeScript": 2012,
    "PHP": 1995,
    "Ruby": 1995,
    "Swift": 2014,
    "Kotlin": 2011,
    "R": 1993,
    "Dart": 2011,
    "MATLAB": 1984,
    "Scala": 2004,
    "Julia": 2012,
    "Perl": 1987,
    "Haskell": 1990,
    "Elixir": 2011,
    "Objective-C": 1984,
    "Assembly": 1949,
    "COBOL": 1959,
    "Fortran": 1957,
    "Ada": 1980,
    "Groovy": 2003,
    "Visual Basic": 1991,
    "Delphi": 1995,
    "Bash": 1989,
    "PowerShell": 2006,
    "Solidity": 2014,
    "Zig": 2016,
    "Crystal": 2014,
    "Nim": 2008
}
```

```python
data = []

for lang, year in languages.items():

    stackoverflow_score = random.randint(40, 100)
    job_demand_score = random.randint(30, 100)
    developer_usage_score = random.randint(30, 100)

    popularity_rating = round(
        (stackoverflow_score + job_demand_score + developer_usage_score) / 30, 2
    )

    data.append({
        "Language": lang,
        "Year_Introduced": year,
        "StackOverflow_Popularity": stackoverflow_score,
        "Job_Demand_Index": job_demand_score,
        "Developer_Usage_Score": developer_usage_score,
        "Popularity_Rating": popularity_rating
    })
```

```python
# Create dataframe
df = pd.DataFrame(data)

# Rank languages
df["Rank"] = df["Popularity_Rating"].rank(ascending=False)

# Sort by rating
df = df.sort_values(by="Popularity_Rating", ascending=False)

# Save dataset
df.to_csv("programming_language_popularity_dataset.csv", index=False)

print("Dataset created successfully")
```
